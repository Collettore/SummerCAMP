```python
#    
import os

path = 'C:\\Gamez\\'

folders = []

# r=root, d=directories, f = files
for r, d, f in os.walk(path):
    for folder in d:
        folders.append(os.path.join(r, folder))

for f in folders:
    print(f)
```

```python
#      
import glob

path = 'C:\\Gamez\\'

files = [f for f in glob.glob(path + "**/*", recursive=True)]

for f in files:
    print(f)
```
